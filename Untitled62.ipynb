```python
#required 
import pandas as pd
from gensim.summarization.summarizer import summarize  #Parameters= text,ratio,word_count,split
from gensim.summarization import keywords #  keywords(text)
from nltk.tokenize import word_tokenize #tokenize 
from nltk.corpus import stopwords #stop words
from nltk.stem.wordnet import WordNetLemmatizer # converts ...rocks=rock, better=good, corpora=corpus
import gensim #topic modeling
from gensim.utils import simple_preprocess #converts a document into list of tokens
import pyLDAvis
import pyLDAvis.gensim #topic model visualization(lda)
import matplotlib.pyplot as plt
%matplotlib inline
import re
import requests
import nltk 
import urllib #acces the internet
from bs4 import BeautifulSoup # Web Scraping tool 
from gensim.models import CoherenceModel
```

```python
#read a file and convert into str
with open('/Users/apple/Desktop/text.txt', 'r') as file:
    text = file.read().replace('\n', '')
```

```python
# returns the summary of the text with three parameters 
#1) given text  2) ratio numbers between 0 to 1   3) split, If True, list of sentences will be returned.
def get_summary(text, pct):
    summary = summarize(text,ratio=pct,split=True) 
    return summary
''''#get most ranked words of provided texts
#ratio parameter the number of sentences is reduced by the provided ratio'''
def get_keywords(text):
    res = keywords(text, ratio=0.1, words=None, split=False, scores=False, pos_filter=('NN', 'JJ'), lemmatize=False, deacc=False)
    res = res.split('\n')
    return res
''''#Convert a document into a list of lowercase tokens, ignoring tokens that are too short or too long
#Tokenize the sentence into words & remove punctuation'''
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True)) #deacc
        
#refere this       
def split_sentences(text):
    """ Split text into sentences.
    """
    sentence_delimiters = re.compile(u'[\\[\\]\n.!?]')
    sentences = sentence_delimiters.split(text)
    return sentences
def split_into_tokens(text):
    """ Split text into tokens.
    """
    #returns partsofspeach
    tokens = nltk.word_tokenize(text)
    return tokens
#return title for text using urllib
def extract_title_text(url):
    page = urllib.request.urlopen(url).read().decode('utf8')
    soup = BeautifulSoup(page,'lxml')
    text = ' '.join(map(lambda p: p.text, soup.find_all('p')))
    return soup.title.text, text        
```

```python
print(get_summary(text, 0.2)) #ratio will decide the size of summary(2nd parameter)
```

 print(get_keywords(text))

```python
"preprocessing and cleaning" 
import re
new_text=re.sub('[^ a-zA-Z0-9]','',text)
#new_text
stop_words=set(stopwords.words('english'))
lema=WordNetLemmatizer()
tokens=word_tokenize(str(new_text))
filtered_sentence = [w for w in tokens if not w in stop_words]
normalized = " ".join(lema.lemmatize(word) for word in filtered_sentence)
normalized
```

```python
from nltk import ngrams
n = 7
total_grams = []
number_of_grams = ngrams(normalized.split(), n)
for grams in number_of_grams:
    total_grams.append(grams)
total_grams[:10]
```

```python
'WORD CLOUD  VISUALIZATION'
from wordcloud import WordCloud
wordcloud = WordCloud(max_font_size=60).generate(normalized)
plt.figure(figsize=(16,12))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
count = {}
for w in normalized.split():
    if w in count:
        count[w] += 1
    else:
        count[w] = 1
for word, times in count.items():
    if times > 3:
        print("%s was found %d times" % (word, times))
```

```python
'Topic Modeling (LDA)'
import gensim
from gensim import corpora
tokenized_sents = list(sent_to_words(filtered_sentence)) #Filtered_sentence= tokenize sentence and remove punctuations ..
dictionary = corpora.Dictionary(tokenized_sents)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in tokenized_sents]
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix, num_topics=2, id2word = dictionary, passes=100)
print(ldamodel.print_topics(num_topics=2, num_words=20))

```

```python
'''Compute Perplexity'''

# a measure of how good the model is. lower the better.

print('\nPerplexity: ', ldamodel.log_perplexity(doc_term_matrix))
coherence_model_lda = CoherenceModel(model=ldamodel, texts=tokenized_sents, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```
